```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Suponiendo que df es tu DataFrame
X = train.drop('label', axis=1)  # Features
y = train['label']  # Variable dependiente

# Crear y entrenar el modelo de regresiÃ³n logÃ­stica multinomial
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
model.fit(X, y)

# Obtener los coeficientes de cada variable
coeficients = model.coef_
intercept = model.intercept_

# Imprimir los coeficientes de las variables
coeficients_df = pd.DataFrame(coeficients, columns=X.columns)
intercept_df = pd.DataFrame(intercept, columns=['intercept'])
result_df = pd.concat([intercept_df, coeficients_df], axis=1)
print("Coeficients:")
print(result_df)
```

```python
# Calcular la matriz de correlaciÃ³n
correlation_matrix = train.corr()

# Imprimir la matriz de correlaciÃ³n
print("Matriz de correlaciÃ³n:")
print(correlation_matrix)
```

```python
correlation_matrix['label']
```

```python
dict_cor = correlation_matrix['label'].to_dict()
```

```python
dict_cor
```

```python
for index, row in train_cor.iterrows():
    score = 0.0
    for col in train_cor.columns[:-1]:
        if col in dict_cor: 
            if row[col] == 1:
                score += dict_cor[col] 
    train_cor.at[index, 'score_by_correlation'] = score
```

```python
train_cor
```

```python
for index, row in test_cor.iterrows():
    score = 0.0  
    for col in test_cor.columns[:-1]:  
        if col in dict_cor: 
            if row[col] == 1: 
                score += dict_cor[col]  
    test_cor.at[index, 'score_by_correlation'] = score
```

```python
test_cor
```

```python
from sklearn.model_selection import train_test_split

X4 = train_cor.drop('label', axis=1)  # Todas las columnas excepto la columna objetivo
Y4 = train_cor['label']

# Dividimos el conjunto de entrenamiento en nuevo entrenamiento y validaciÃ³n (85% entrenamiento, 15% validaciÃ³n del conjunto de entrenamiento original)
X4_train, X4_val, Y4_train, Y4_val = train_test_split(X4, Y4, test_size=0.125, random_state=42)
```

```python
from catboost import CatBoostClassifier

best_hyperparams = {'depth': 8, 'iterations': 30, 'l2_leaf_reg': 1, 'learning_rate': 0.01}
best_cb_model = CatBoostClassifier(**best_hyperparams)
best_cb_model.fit(X4_train, Y4_train, verbose=False)

feature_importances = best_cb_model.get_feature_importance()
feature_names = X4_train.columns

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

importance_df = importance_df.sort_values(by='Importance', ascending=False)


test_end4_sin_label = test_cor.drop('label', axis=1)
predicciones4 = best_cb_model.predict(test_end4_sin_label)
probabilidades4 = best_cb_model.predict_proba(test_end4_sin_label)

predicted_df4 = test_cor.copy()
predicted_df4['label'] = predicciones4

print(predicted_df4.head())

```

```python
# Restablecer el Ã­ndice de predicted_df para que comience en 0 y luego sumar 1 para comenzar desde 1
predicted_df4.reset_index(drop=True, inplace=True)
predicted_df4['ID'] = predicted_df4.index 

# Seleccionar las columnas 'ID' y 'label' para el nuevo DataFrame df_randomf
df_randomf3 = predicted_df4[['ID', 'label']]

# Guardar df_randomf en un archivo CSV sin el Ã­ndice de pandas
df_randomf3.to_csv('try69_data.csv', index=False)
```
