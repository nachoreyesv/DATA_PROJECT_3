```python
train = pd.read_parquet('train_best.parquet')
```

```python
test = pd.read_parquet('test_best.parquet')
```

```python
train['vars_num'] = train['age'] + train['trestbps'] + train['chol'] + train['thalach'] + train['oldpeak']
```

```python
train
```

```python
test['vars_num'] = test['age'] + test['trestbps'] + test['chol'] + test['thalach'] + test['oldpeak'] 
```

```python
test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer

target_column = 'label'
features_train = train.drop(columns=[target_column])
target_train = train[target_column]

features_test = test.drop(columns=[target_column])

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(features_train, target_train)

predicted_classes = model.predict(features_test)

predicted_df_cor = test.copy()
predicted_df_cor[target_column] = predicted_classes
```

```python
predicted_df_cor
```

```python
predicted_df_cor['ID'] = range(len(predicted_df_cor))
predicted_df_cor[['ID', 'label']].to_csv('tryTEST_data.csv', index=False)
```
